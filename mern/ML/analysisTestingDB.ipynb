```python
from pymongo import MongoClient
import os
import pymongo
import pandas as pd
import requests

# Create a client
# client = MongoClient('mongodb://localhost:27017/')
# Read the connection string from the config.env file
with open('../server/config.env', 'r') as file:
    connection_string = file.read().strip()

connection_string = connection_string.replace('ATLAS_URI=', '')
# print(connection_string)

# Create a client
client = pymongo.MongoClient(connection_string)
```

```python
import pandas as pd

# Connect to your database
db = client['sample_analytics']

# Get your collection
collection = db['transactions']

# Retrieve all documents
documents = collection.find()

# Convert documents into a DataFrame
df = pd.DataFrame(list(documents))

# Explode the transactions column
df = df.explode('transactions')

# Convert the dictionaries in the transactions column into separate columns
df_transactions = df['transactions'].apply(pd.Series)

# Concatenate the original DataFrame with the new transactions DataFrame
df = pd.concat([df.drop(['transactions'], axis=1), df_transactions], axis=1)

df.head(20)
```

```python
import os
import pandas as pd

# Specify the path to your directory
dir_path = "C:/Github/FYP/Data/stocks"

# Get a list of all files in the directory
files = os.listdir(dir_path)

# Initialize a counter
counter = 0

# Loop over all files
for file in files:
    # Break the loop after 3 iterations
    if counter == 3:
        break

    # Construct the full file path
    file_path = os.path.join(dir_path, file)
    
    # Read the file
    df = pd.read_csv(file_path)
    
    # Extract the ticker from the file name (remove .txt)
    ticker = os.path.splitext(file)[0]
    
    # Remove .us from the ticker name
    ticker = ticker.replace('.us', '')
    
    # Add a 'Ticker' column to the DataFrame
    df['Ticker'] = ticker
    
    # Print the DataFrame
    print(df)

    # Increment the counter
    counter += 1
```
